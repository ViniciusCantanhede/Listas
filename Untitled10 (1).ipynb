```python
#Questão 8
import sympy as sp

# Definindo as variáveis M e C
M, C = sp.symbols('M C')

# Definindo a função do QI
QI = (M / C) * 100

# Calculando as derivadas parciais
dQI_dM = sp.diff(QI, M)
dQI_dC = sp.diff(QI, C)

# Substituindo o ponto (12, 10)
dQI_dM_value = dQI_dM.subs([(M, 12), (C, 10)])
dQI_dC_value = dQI_dC.subs([(M, 12), (C, 10)])

print(f"A derivada parcial ∂QI/∂M no ponto (12, 10) é: {dQI_dM_value}")
print(f"A derivada parcial ∂QI/∂C no ponto (12, 10) é: {dQI_dC_value}")
```

```python


#Questão 2, lista 9
import math

# Define a função de temperatura
def temperatura(x, y):
    return 4 * x**2 - 4 * x * y + y**2

# Inicializar temperaturas máxima e mínima
temperatura_maxima = float('-inf')
temperatura_minima = float('inf')

# Percorrer o círculo de raio 5
for angulo in range(0, 360):
    # Converter ângulo para radianos
    radiano = math.radians(angulo)
    
    # Calcular as coordenadas x e y no círculo de raio 5
    x = 5 * math.cos(radiano)
    y = 5 * math.sin(radiano)
    
    # Calcular a temperatura no ponto (x, y)
    temp = temperatura(x, y)
    
    # Atualizar temperatura máxima e mínima
    if temp > temperatura_maxima:
        temperatura_maxima = temp
    if temp < temperatura_minima:
        temperatura_minima = temp

# Imprimir os resultados
print("Maior temperatura encontrada pela formiga:", temperatura_maxima)
print("Menor temperatura encontrada pela formiga:", temperatura_minima)


```

```python
#Grafico para demonstração 
import numpy as np
import matplotlib.pyplot as plt

# Define a função de temperatura
def temperatura(x, y):
    return 4 * x**2 - 4 * x * y + y**2

# Inicializar ângulos
angulos = np.linspace(0, 2*np.pi, 100)

# Inicializar temperaturas
temperaturas = []

# Calcular as temperaturas ao longo do círculo
for angulo in angulos:
    x = 5 * np.cos(angulo)
    y = 5 * np.sin(angulo)
    temp = temperatura(x, y)
    temperaturas.append(temp)

# Plotar o gráfico
plt.plot(angulos, temperaturas)
plt.title("Distribuição da temperatura ao longo do círculo")
plt.xlabel("Ângulo (em radianos)")
plt.ylabel("Temperatura")
plt.grid(True)
plt.show()
```
